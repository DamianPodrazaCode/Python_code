```python
car_sales = pd.read_csv('car-sales.csv')
```

```python
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('$', '').str.replace(',', '').str.replace('.00', '')
# car_sales['Price'] = car_sales['Price'].str[:-2] # jeżeli wcześniej nie byłyby usunięte ostatnie dwa zera, to tu usuwa ostatnie dwa znaki
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype(int) # zamiana z str na int
car_sales['Sale Date'] = pd.date_range('1/1/2020', periods=len(car_sales)) # dodanie kolumny
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales
```

```python
car_sales.plot(x='Sale Date', y='Total Sales')
;
```

```python
car_sales.plot(x='Odometer (KM)', y='Price', kind='bar')
;
```

```python
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter')
;
```
