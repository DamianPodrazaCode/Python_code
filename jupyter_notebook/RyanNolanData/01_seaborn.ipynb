```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns.get_dataset_names() # nazwy przykładowych danych
```

```python
mpg = sns.load_dataset('mpg')
mpg
```

```python
sns.set_theme() # Ustawia aspekty motywu wizualnego dla wszystkich wykresów matplotlib i seaborn.
```

```python
# Ustaw parametry kontrolujące ogólny styl wykresów.
# style= {darkgrid, whitegrid, dark, white, ticks}
sns.set_style(style='whitegrid') 
```

```python
# Ustawienie parametrów kontrolujących skalowanie elementów wykresu.
# context= {paper, notebook, talk, poster} -> predefiniowane wielkości
# font_scale= -> zmiana skali fonta od od predefiniowanego contekstu
sns.set_context(context='paper', font_scale=0.8)
```

```python
sns.scatterplot(data=mpg, x='weight', y='mpg')
```

```python
# Oblicza korelację parami kolumn, z wyłączeniem wartości NA/null.
# Jeżeli jedna i druga kolumna rośnie liniowo to jest dora dodatnia korelacja i wynik zbliża się do 1
# Jeżeli jedna kolumna rośnie a druga maleje liniowo to jest dora ujemna korelacja i wynik zbliża się do -1
# Jeżeli wyniki nie zmieniają się wspólnie to wyniki są zbliżające siędo 0
mpg.corr(numeric_only=True)
```

```python
# Wykreśla prostokątne dane jako macierz kodowaną kolorami.
sns.heatmap(mpg.corr(numeric_only=True))
```

```python
# wymiary wykresu
plt.figure(figsize=(10, 5)) 

# hue= rozdziela na kjolory wg. kolumny, palette= wybiera kolorystuke link niżej
# https://seaborn.pydata.org/tutorial/color_palettes.html#choosing-color-palettes
# dla matplotlib https://matplotlib.org/stable/users/explain/colors/colormaps.html
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin', palette='bright') 

# Usuń górny i prawy grzbiet z działek.
sns.despine(left=True, right=False) # true usuwa
```

```python
sns.lineplot(data=mpg, x='model_year', y='mpg', hue='origin', palette='bright')
```

```python
sns.pointplot(data=mpg, x='model_year', y='mpg', hue='origin', palette='bright')
```

```python
sns.kdeplot(data=mpg, x='weight', y='mpg')
```

```python
sns.kdeplot(data=mpg, x='weight', y='mpg', fill=True)
```

```python
sns.histplot(data=mpg, x='weight', y='mpg')
```

```python
sns.histplot(data=mpg, x='mpg')
```

```python
sns.histplot(mpg['mpg'])
```

```python
sns.distplot(mpg['mpg'])
```

```python
sns.histplot(data=mpg, x='mpg', kde=True, bins=20) # bins= ile ma być kolumn na wykresie
```
