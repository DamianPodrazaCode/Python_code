```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [11,22,33,44,55]
```

## plot

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.plot(x, y) # rysowanie z list
ax.plot(x, [1,2,3,4,5])
ax.set(title="Simple Plot", xlabel='x-axis', ylabel='y-axis')
fig.savefig("sample-plt.png")
;
```

## scatter

```python
x = np.linspace(0, 10 ,100) # generuje 100 liczb w równych odstępach od 0 do 10 (float) (start, end, count)
x
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x)) # punkty
;
```

## bar

```python
nut_butter_prices = {'Almond butter': 10,
                     'Penaut butter': 8,
                     'Cashew butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # słupki pionowe
ax.set(title='tytuł', ylabel='y')
;
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())) # słupki poziome
ax.set(title='tytuł', xlabel='x')
;
```
