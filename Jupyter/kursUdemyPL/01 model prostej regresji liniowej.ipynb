# Model prostej regresji liniowej

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
zestaw_danych = pd.read_csv('mieszkania.csv')
X = zestaw_danych.iloc[:,0:1].values
y = zestaw_danych.iloc[:,-1].values
```

```python
# print(zestaw_danych)
# print(type(zestaw_danych))
# print(X)
# print(type(X))
# print(y)
# print(type(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# print(X_test)
# print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# print(y_pred)
# print(y_test)
```

```python
plt.scatter(X_train, y_train, color='red') # dane treningowe 80%
plt.plot(X_train, model.predict(X_train), color='blue') # regresja liniowa wyuczona wg danych treningowych (model) 
plt.scatter(X_test, y_test, color='yellow') # dane testowe 20%
plt.scatter(X_test, y_pred, color='blue') # przewidywania wg danych testowych 
plt.show()
```
