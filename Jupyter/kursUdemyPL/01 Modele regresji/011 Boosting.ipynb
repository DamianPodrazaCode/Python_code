## Boosting

Boosting polega na sekwencyjnym trenowaniu wielu modeli, gdzie każdy kolejny model stara się poprawić błędy poprzednich modeli. Modele są trenowane sekwencyjnie, a każdy kolejny model koryguje błędy poprzedniego.

Cechy:
Zwiększa dokładność modelu poprzez korygowanie błędów poprzednich modeli.

Może prowadzić do bardzo dokładnych modeli.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor

dane = pd.read_csv('004_mieszkania.csv')
X = dane.iloc[:, :1].values # kolumna pierwasza 
y = dane.iloc[:, -1].values # ostatnia kolumna

# Tworzenie modelu Gradient Boosting
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
model.fit(X, y)
```

```python
# Wizualizacja polega na stworzeniu tablicy X-ów na tyle gęstych żeby było widać dokładny przebieg krzywej modelu, (ilość punktów X = max(X)-min(X))
X_in = np.linspace(np.min(X), np.max(X), (np.max(X)-np.min(X))) # wygenerowanie X wejściowych w zakresie danych wejściowych między min(X) - max(X)
X_in = X_in.reshape(-1, 1) # konwersja do tablicy dwuwymiarowej
prognoza = model.predict(X_in) 

plt.scatter(X, y, color='red') 
plt.plot(X_in, prognoza, color='blue') 
plt.show()
```
