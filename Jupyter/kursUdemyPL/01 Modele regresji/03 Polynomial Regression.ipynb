# Regresja wielomianowa

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
```

```python
dane = pd.read_csv('03_mieszkania.csv')
X = dane.iloc[:, 0:1].values # kolumna pierwasza 
y = dane.iloc[:, -1].values # ostatnia kolumna
```

## Model regresji liniowej

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prognoza = model.predict(X)
```

```python
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot(X, prognoza, color='blue') # prognoza 
plt.show()
```

## Model regresji wielomianowej 

```python
# Zamiana danych wejściowych na wielomiany
wiel = PolynomialFeatures(degree=3) # degree -> stopień wielomianu (jak 1 to regresja liniowa) 
X_wiel = wiel.fit_transform(X) # stworzenie danych pomocniczych wielomianowych dla wejścia danych
```

```python
model = LinearRegression() 
model.fit(X_wiel, y)
```

```python
prognoza = model.predict(X_wiel)
```

```python
# wizualizacja uproszczona od danej do danej
plt.scatter(X, y, color='red') 
plt.plot(X, prognoza, color='blue')
plt.show()
```

```python
# przygotowanie danych do dokładnej wizualizacji
X_in = np.arange(np.min(X), np.max(X), 0.1) # wygenerowanie X wejściowych o gęstości co 0.1, w zakresie danych wejściowych
X_in = X_in.reshape(len(X_in), 1) # konwersja do tablicy dwywymiarowej
X_wiel = wiel.fit_transform(X_in) # transformacja na wielomiany
prognoza = model.predict(X_wiel) 
```

```python
# wizualizacja dokładna co 0.1
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot(X_in, prognoza, color='blue') # regresja 
plt.show()
```
