# Regresja wielomianowa

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
```

```python
dane = pd.read_csv('03_mieszkania.csv')
X = dane.iloc[:, 0:1].values # kolumna pierwasza 
y = dane.iloc[:, -1].values # kolumna ostatnia
```

```python
# Zamiana danych wejściowych na wielomiany
wiel = PolynomialFeatures(degree=4) # degree -> stopień wielomianu 
X_wiel = wiel.fit_transform(X) 
```

```python
model = LinearRegression() 
model.fit(X_wiel, y)
```

```python
prognoza = model.predict(X_wiel)
```

```python
# wizualizacja uproszczona od danej do danej
plt.scatter(X, y, color='red') 
plt.plot(X, prognoza, color='blue')
plt.show()
```

```python
# przygotowanie danych do wizualizacji ze zwiększoną liczbą punktów X
X_in = np.arange(np.min(X), np.max(X), 0.1) # wygenerowanie X wejściowych o gęstości co 0.1, w zakresie danych wejściowych
# X_in = np.linspace(np.min(X), np.max(X), 100) # lub wygenerowanie 100 punktów między min a mix
X_in = X_in.reshape(-1, 1) # konwersja do tablicy dwuwymiarowej
# transformacja na wielomiany, ponieważ model był uczony na wielomianach, to prognoza też musi być obliczna na wielomianach
X_wiel = wiel.fit_transform(X_in) 
prognoza = model.predict(X_wiel) 
```

```python
# wizualizacja dokładna
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot(X_in, prognoza, color='blue') # regresja 
plt.show()
```
