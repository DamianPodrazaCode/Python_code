# Regresja wielomianowa dla wielu cech

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
```

```python
# dane = pd.read_csv('03_mieszkania.csv')
dane = pd.read_csv('02_mieszkania.csv')
X = dane.iloc[:, :-1].values # kolumny oprócz ostatniej  
y = dane.iloc[:, -1].values # kolumna ostatnia
```

```python
# Zamiana danych wejściowych na wielomiany
wiel = PolynomialFeatures(degree=4) # degree -> stopień wielomianu 
X_wiel = wiel.fit_transform(X) 
```

```python
model = LinearRegression() 
model.fit(X_wiel, y)
```

```python
prognoza = model.predict(X_wiel)
```

```python
porownanie = pd.DataFrame(X, columns=['m2','pokoje','rok'])
porownanie['Cena rzeczywista'] = y
porownanie['Cena przewidywana'] = np.round(prognoza, 0)
porownanie['Błąd %'] = np.round(((porownanie['Cena rzeczywista'] - porownanie['Cena przewidywana']) / porownanie['Cena rzeczywista']) * 100, 0)
print(porownanie)
```

```python
# prognozy dla jednego rekordu danych wejściowech
X_in = [[70, 3, 2021]]
X_in = wiel.fit_transform(X_in) 
prognoza = model.predict(X_in)
print(X_in, prognoza)
```
