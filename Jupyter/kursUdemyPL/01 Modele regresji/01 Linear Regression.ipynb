# Regresja Liniowa

Nazwewnictwo:
- X -> Features (cechy, zmienne), Attributes (atrybuty)
- y (w postaci wejściowej) -> Labels (etykiety)
- y (w postaci prognozy) -> Target (wartość docelowa)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
dane = pd.read_csv('04_mieszkania.csv') 
X = dane.iloc[:, 0:1].values # wyekstrahowanie kolumny pierwszej z metrażem (format -> [[0],[1],[2], ...])
y = dane.iloc[:, -1].values # wyekstrahowanie ostatniej kolumny (format -> [0, 1, 2, ...])
```

```python
model = LinearRegression() # stworzenie modelu treningowego
model.fit(X, y) # trening - dopasowanie danych
```

```python
prognoza = model.predict(X) # stworzenie tablicy prognozy wg modelu dla wszystkich X
```

```python
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot([np.min(X), np.max(X)], [np.min(prognoza), np.max(prognoza)] , color='blue') # linia regresji (prognozy), pomiędzy min a max punktem
# model jest linią prostą, więc wystarczy narysować linie pomiędzy punktami min i max
plt.show()
```

```python
# prognozy dla jednej danej wejściowej
X_in = [[95]] # można naraz wyłuskać więcej danych np.[[21],[76],[34]]
prognoza = model.predict(X_in)
print(X_in, prognoza)
```
