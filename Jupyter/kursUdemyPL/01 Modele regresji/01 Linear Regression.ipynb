# Regresja Liniowa

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
dane = pd.read_csv('01_mieszkania.csv')
X = dane.iloc[:, 0:1].values # wyekstrahowanie kolumny pierwszej z metrażem (format -> [[0],[1],[2], ...])
y = dane.iloc[:, -1].values # wyekstrahowanie ostatniej kolumny (format -> [0, 1, 2, ...])
```

```python
model = LinearRegression() # stworzenie modelu treningowego
model.fit(X, y) # trening - dopasowanie danych
```

```python
prognoza = model.predict(X) # stworzenie tablicy prognozy wg modelu
```

```python
plt.scatter(X, y, color='red') # dane wejściowe
plt.scatter(X, prognoza, color='blue') # prognoza
plt.plot(X, prognoza, color='black') # linia regresji (prognozy)
plt.show() 
```

```python
# sposób wyliczenia jednej danej wg modelu
X_in = [[271]] 
y_out = model.predict(X_in)
print(X_in, y_out)
```
