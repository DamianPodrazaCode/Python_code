# Drzewo decyzyjne regresja, multi cecha

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

dane = pd.read_csv('01_mieszkania.csv')
X = dane.iloc[:, :-1].values
y = dane.iloc[:, -1].values 
y = y.reshape(-1, 1) 

X_tren, X_test, y_tren, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = DecisionTreeRegressor(random_state=0)
model.fit(X_tren, y_tren)
```

```python
prognoza = model.predict(X_test)

# wizualizacja uproszczona od danej do danej
plt.scatter(X_tren[:, 0], y_tren, color='red') 
plt.scatter(X_test[:, 0], y_test, color='green') 
plt.scatter(X_test[:, 0], prognoza, color='blue')

plt.show()
```
