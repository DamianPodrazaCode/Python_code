# Drzewo decyzyjne regresja

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
dane = pd.read_csv('04_mieszkania.csv')
X = dane.iloc[:, :1].values # kolumna pierwasza 
y = dane.iloc[:, -1].values # ostatnia kolumna
```

```python
model = DecisionTreeRegressor(random_state=0)
model.fit(X, y)
```

```python
prognoza = model.predict(X)
```

```python
# wizualizacja uproszczona od danej do danej
plt.scatter(X, y, color='red') 
plt.plot(X, prognoza, color='blue')
plt.show()
```

```python
# przygotowanie danych do wizualizacji ze zwiększoną liczbą punktów X
X_in = np.arange(np.min(X), np.max(X), 0.1) # wygenerowanie X wejściowych o gęstości co 0.1, w zakresie danych wejściowych
# X_in = np.linspace(np.min(X), np.max(X), 100) # lub wygenerowanie 100 punktów między min a mix
X_in = X_in.reshape(-1, 1) # konwersja do tablicy dwuwymiarowej
prognoza = model.predict(X_in) 
```

```python
# wizualizacja dokładna 
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot(X_in, prognoza, color='blue') # regresja 
plt.show()
```
