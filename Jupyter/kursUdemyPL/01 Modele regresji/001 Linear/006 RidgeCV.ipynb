RidgeCV to rozszerzenie Ridge Regression, które wprowadza cross-validation (walidację krzyżową) w celu automatycznego wyboru najlepszego parametru regularyzacji (𝛼). Dzięki temu model może dostosować się do danych i wybrać optymalną wartość 𝛼 zamiast korzystać z ustalonej wartości.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
```

```python
dane = pd.read_csv('../DataIn/004_mieszkania.csv') 
X = dane.iloc[:, 0:1].values # wyekstrahowanie kolumny pierwszej z metrażem (format -> [[0],[1],[2], ...])
y = dane.iloc[:, -1].values # wyekstrahowanie ostatniej kolumny z ceną (format -> [0, 1, 2, ...])
```

```python
model = RidgeCV(
    alphas=(0.1, 1.0, 10.0),
    fit_intercept=True,
    scoring=None,
    cv=None,
    gcv_mode=None,
    store_cv_results=None,
    alpha_per_target=False,
    store_cv_values='deprecated'
    )
model.fit(X, y) 
```

```python
prognoza = model.predict(X) # stworzenie tablicy prognozy wg modelu dla wszystkich X
```

```python
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot([np.min(X), np.max(X)], [np.min(prognoza), np.max(prognoza)] , color='blue') # linia regresji (prognozy), pomiędzy min a max punktem
# model jest linią prostą, więc wystarczy narysować linie pomiędzy punktami min i max
plt.show()
```
