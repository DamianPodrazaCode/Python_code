RidgeCV to rozszerzenie Ridge Regression, ktÃ³re wprowadza cross-validation (walidacjÄ™ krzyÅ¼owÄ…) w celu automatycznego wyboru najlepszego parametru regularyzacji (ð›¼). DziÄ™ki temu model moÅ¼e dostosowaÄ‡ siÄ™ do danych i wybraÄ‡ optymalnÄ… wartoÅ›Ä‡ ð›¼ zamiast korzystaÄ‡ z ustalonej wartoÅ›ci.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
```

```python
dane = pd.read_csv('../DataIn/004_mieszkania.csv') 
X = dane.iloc[:, 0:1].values # wyekstrahowanie kolumny pierwszej z metraÅ¼em (format -> [[0],[1],[2], ...])
y = dane.iloc[:, -1].values # wyekstrahowanie ostatniej kolumny z cenÄ… (format -> [0, 1, 2, ...])
```

```python
model = RidgeCV(
    alphas=(0.1, 1.0, 10.0),
    fit_intercept=True,
    scoring=None,
    cv=None,
    gcv_mode=None,
    store_cv_results=None,
    alpha_per_target=False,
    store_cv_values='deprecated'
    )
model.fit(X, y) 
```

```python
prognoza = model.predict(X) # stworzenie tablicy prognozy wg modelu dla wszystkich X
```

```python
plt.scatter(X, y, color='red') # dane wejÅ›ciowe
plt.plot([np.min(X), np.max(X)], [np.min(prognoza), np.max(prognoza)] , color='blue') # linia regresji (prognozy), pomiÄ™dzy min a max punktem
# model jest liniÄ… prostÄ…, wiÄ™c wystarczy narysowaÄ‡ linie pomiÄ™dzy punktami min i max
plt.show()
```
