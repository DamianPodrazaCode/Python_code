Regresja liniowa z większą możliwością regulacji.
Ridge Regression, znana również jako regresja grzbietowa lub regularyzacja L2, jest techniką regresji liniowej, która wprowadza regularyzację w celu poprawy modelu regresji i zapobiegania przeuczeniu (overfitting). Ridge Regression dodaje karę (penalty) za wielkość współczynników regresji do funkcji kosztu.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
```

```python
dane = pd.read_csv('../DataIn/004_mieszkania.csv') 
X = dane.iloc[:, 0:1].values # wyekstrahowanie kolumny pierwszej z metrażem (format -> [[0],[1],[2], ...])
y = dane.iloc[:, -1].values # wyekstrahowanie ostatniej kolumny z ceną (format -> [0, 1, 2, ...])
```

```python
model = Ridge(
    alpha=1.0,
    fit_intercept=True,
    copy_X=True,
    max_iter=None,
    tol=0.0001,
    solver='auto',
    positive=False,
    random_state=None
    )
model.fit(X, y) 
```

```python
prognoza = model.predict(X) # stworzenie tablicy prognozy wg modelu dla wszystkich X
```

```python
plt.scatter(X, y, color='red') # dane wejściowe
plt.plot([np.min(X), np.max(X)], [np.min(prognoza), np.max(prognoza)] , color='blue') # linia regresji (prognozy), pomiędzy min a max punktem
# model jest linią prostą, więc wystarczy narysować linie pomiędzy punktami min i max
plt.show()
```
