## 2.1 Listy

### Tworzenie list

```python
# === TWORZENIE LIST ===

# Lista pusta
lista_pusta = []
lista_pusta2 = list()

# Lista z elementami
liczby = [1, 2, 3, 4, 5]
owoce = ["jabłko", "banan", "czereśnia"]
mieszana = [1, "tekst", 3.14, True, None, [1, 2]]

# Tworzenie z range()
od_0_do_9 = list(range(10))
parzyste = list(range(0, 20, 2))

# Tworzenie przez powtórzenie
zera = [0] * 10
wzorzec = [1, 2, 3] * 3

# Tworzenie z ciągu znaków
litery = list("Python")

# Tworzenie z funkcji split()
slowa = "Ala ma kota".split()

print(f"liczby:     {liczby}")
print(f"owoce:      {owoce}")
print(f"mieszana:   {mieszana}")
print(f"od_0_do_9:  {od_0_do_9}")
print(f"parzyste:   {parzyste}")
print(f"zera:       {zera}")
print(f"wzorzec:    {wzorzec}")
print(f"litery:     {litery}")
print(f"slowa:      {slowa}")
```
