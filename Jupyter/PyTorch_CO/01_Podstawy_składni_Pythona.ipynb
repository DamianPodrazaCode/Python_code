## 1.1 Zmienne i typy danych
### Czym jest zmienna?
Zmienna to nazwany kontener przechowujƒÖcy dane w pamiƒôci komputera. W Pythonie nie trzeba deklarowaƒá typu zmiennej ‚Äî interpreter rozpoznaje go automatycznie (typowanie dynamiczne).

```python
# ============================================
# KOM√ìRKA 1: Tworzenie zmiennych
# ============================================

# Python nie wymaga deklaracji typu - sam go rozpoznaje
wiek = 25
nazwa = "Python"
temperatura = 36.6
czy_programista = True

print("Zmienna 'wiek' =", wiek)
print("Zmienna 'nazwa' =", nazwa)
print("Zmienna 'temperatura' =", temperatura)
print("Zmienna 'czy_programista' =", czy_programista)
```

### Typ int (liczby ca≈Çkowite)

```python
# ============================================
# KOM√ìRKA 2: Typ int - liczby ca≈Çkowite
# ============================================

# Podstawowe liczby ca≈Çkowite
wiek = 25
rok_urodzenia = 1999
temperatura_minus = -15
zero = 0

print("=== TYP INT (liczby ca≈Çkowite) ===")
print(f"wiek = {wiek}, typ: {type(wiek)}")
print(f"rok_urodzenia = {rok_urodzenia}, typ: {type(rok_urodzenia)}")
print(f"temperatura_minus = {temperatura_minus}, typ: {type(temperatura_minus)}")
print(f"zero = {zero}, typ: {type(zero)}")

print("\n--- R√≥≈ºne systemy liczbowe ---")
# Python obs≈Çuguje r√≥≈ºne systemy liczbowe
binarny = 0b1010          # system dw√≥jkowy (binarny)
osemkowy = 0o17            # system √≥semkowy (oktalny)
szesnastkowy = 0xFF        # system szesnastkowy (heksadecymalny)

print(f"0b1010 (binarny) = {binarny}")
print(f"0o17 (√≥semkowy) = {osemkowy}")
print(f"0xFF (szesnastkowy) = {szesnastkowy}")

print("\n--- Du≈ºe liczby z separatorem ---")
# Podkre≈õlnik jako separator tysiƒôcy (czytelno≈õƒá)
populacja_polski = 38_000_000
budzet = 1_500_000_000

print(f"Populacja Polski: {populacja_polski}")
print(f"Bud≈ºet: {budzet}")

print("\n--- Python obs≈Çuguje DOWOLNIE du≈ºe liczby! ---")
bardzo_duza = 10 ** 100  # googol
print(f"10^100 = {bardzo_duza}")
print(f"Liczba cyfr: {len(str(bardzo_duza))}")
```

### Typ float (liczby zmiennoprzecinkowe)

```python
# ============================================
# KOM√ìRKA 3: Typ float - liczby zmiennoprzecinkowe
# ============================================

print("=== TYP FLOAT (liczby zmiennoprzecinkowe) ===\n")

# Podstawowe floaty
pi = 3.14159
temperatura = -5.5
wzrost = 1.75
cena = 19.99

print(f"pi = {pi}, typ: {type(pi)}")
print(f"temperatura = {temperatura}, typ: {type(temperatura)}")
print(f"wzrost = {wzrost}, typ: {type(wzrost)}")
print(f"cena = {cena}, typ: {type(cena)}")

print("\n--- Notacja naukowa (wyk≈Çadnicza) ---")
# Notacja naukowa: e oznacza "razy 10 do potƒôgi"
predkosc_swiatla = 3e8       # 3 √ó 10^8 = 300 000 000
masa_elektronu = 9.109e-31   # 9.109 √ó 10^(-31)
avogadro = 6.022e23          # 6.022 √ó 10^23

print(f"Prƒôdko≈õƒá ≈õwiat≈Ça: {predkosc_swiatla} m/s")
print(f"Masa elektronu: {masa_elektronu} kg")
print(f"Liczba Avogadra: {avogadro}")

print("\n--- Specjalne warto≈õci float ---")
nieskonczonosc = float('inf')
minus_nieskonczonosc = float('-inf')
nie_liczba = float('nan')

print(f"Niesko≈Ñczono≈õƒá: {nieskonczonosc}")
print(f"Minus niesko≈Ñczono≈õƒá: {minus_nieskonczonosc}")
print(f"NaN (Not a Number): {nie_liczba}")
print(f"Czy inf > 1000000? {nieskonczonosc > 1000000}")

print("\n--- ‚ö†Ô∏è UWAGA: Niedok≈Çadno≈õƒá float! ---")
# To jest WA≈ªNE do zapamiƒôtania!
wynik = 0.1 + 0.2
print(f"0.1 + 0.2 = {wynik}")
print(f"0.1 + 0.2 == 0.3? {wynik == 0.3}")  # False!
print(f"R√≥≈ºnica: {wynik - 0.3}")

# Jak sobie z tym radziƒá?
import math
print(f"\nmath.isclose(0.1 + 0.2, 0.3)? {math.isclose(0.1 + 0.2, 0.3)}")
print(f"round(0.1 + 0.2, 1) == 0.3? {round(0.1 + 0.2, 1) == 0.3}")
```

### Typ str (≈Ça≈Ñcuchy znak√≥w / napisy)

```python
# ============================================
# KOM√ìRKA 4: Typ str - ≈Ça≈Ñcuchy znak√≥w
# ============================================

print("=== TYP STR (≈Ça≈Ñcuchy znak√≥w) ===\n")

# R√≥≈ºne sposoby tworzenia string√≥w
print("--- Tworzenie string√≥w ---")

# Apostrofy i cudzys≈Çowy - dzia≈ÇajƒÖ tak samo
imie = 'Anna'
nazwisko = "Kowalska"
print(f"Apostrofy: '{imie}', Cudzys≈Çowy: \"{nazwisko}\"")

# Kiedy u≈ºyƒá kt√≥rych?
zdanie1 = "It's a beautiful day"      # apostrof w tre≈õci ‚Üí cudzys≈Çowy
zdanie2 = 'Powiedzia≈Ç "cze≈õƒá"'        # cudzys≈Çowy w tre≈õci ‚Üí apostrofy
print(f"zdanie1: {zdanie1}")
print(f"zdanie2: {zdanie2}")

# String wielolinijkowy - potr√≥jne cudzys≈Çowy
wiersz = """Roses are red,
Violets are blue,
Python is awesome,
And so are you."""

print(f"\nWiersz:\n{wiersz}")

print("\n--- Znaki specjalne (sekwencje ucieczki) ---")
print("Nowa linia: Linia 1\nLinia 2")
print("Tabulator: Kolumna1\tKolumna2\tKolumna3")
print("Backslash: C:\\Users\\Anna\\Documents")
print("Cudzys≈Ç√≥w: Powiedzia≈Ç \"cze≈õƒá\"")

# Raw string - ignoruje znaki specjalne
print(r'C:\Users\new_folder\test') # tak mo≈ºna ale f-stringu nie mo≈ºna tego u≈ºyƒá
raw_path = r'C:\Users\new_folder\test'
print(f"\nRaw string: {raw_path}")

print("\n--- Operacje na stringach ---")
tekst = "Python"
print(f"Tekst: '{tekst}'")
print(f"D≈Çugo≈õƒá (len): {len(tekst)}")
print(f"Wielkie litery: {tekst.upper()}")
print(f"Ma≈Çe litery: {tekst.lower()}")
print(f"Pierwsza wielka: {'hello world'.capitalize()}")
print(f"Ka≈ºde s≈Çowo wielkƒÖ: {'hello world'.title()}")

print("\n--- Indeksowanie (dostƒôp do znak√≥w) ---")
tekst = "PYTHON"
print(f"Tekst: '{tekst}'")
print(f"Indeksy: P=0, Y=1, T=2, H=3, O=4, N=5")
print(f"tekst[0] = '{tekst[0]}'   (pierwszy znak)")
print(f"tekst[2] = '{tekst[2]}'   (trzeci znak)")
print(f"tekst[-1] = '{tekst[-1]}'  (ostatni znak)")
print(f"tekst[-2] = '{tekst[-2]}'  (przedostatni znak)")

print("\n--- Wycinanie (slicing) ---")
tekst = "Hello, World!"
print(f"Tekst: '{tekst}'")
print(f"tekst[0:5] = '{tekst[0:5]}'      (znaki 0-4)")
print(f"tekst[7:] = '{tekst[7:]}'       (od 7 do ko≈Ñca)")
print(f"tekst[:5] = '{tekst[:5]}'       (od poczƒÖtku do 4)")
print(f"tekst[::2] = '{tekst[::2]}'     (co drugi znak)")
print(f"tekst[::-1] = '{tekst[::-1]}'  (odwr√≥cony)")

print("\n--- ≈ÅƒÖczenie i powielanie ---")
powitanie = "Cze≈õƒá" + " " + "≈öwiecie!"    # konkatenacja
print(f"Konkatenacja: {powitanie}")
linia = "-" * 30                           # powielanie
print(f"Powielanie: {linia}")
print(f"'ha' * 3 = {'ha' * 3}")

print("\n--- Przydatne metody ---")
email = "  Anna.Kowalska@Email.com  "
print(f"Oryginalny: '{email}'")
print(f"strip(): '{email.strip()}'")
print(f"lower(): '{email.strip().lower()}'")
print(f"replace(): '{email.strip().replace('@', ' [at] ')}'")
print(f"split('@'): {email.strip().split('@')}")
print(f"startswith('  A'): {email.startswith('  A')}")
print(f"'@' in email: {'@' in email}")

zdanie = "Python jest super"
print(f"\n'{zdanie}'.split() = {zdanie.split()}")
slowa = ['Python', 'jest', 'super']
print(f"' '.join({slowa}) = {' '.join(slowa)}")
```

### Typ bool (warto≈õci logiczne)

```python
# ============================================
# KOM√ìRKA 5: Typ bool - warto≈õci logiczne
# ============================================

print("=== TYP BOOL (warto≈õci logiczne) ===\n")

# Tylko dwie warto≈õci: True i False (z wielkiej litery!)
jest_studentem = True
ma_dlug = False

print(f"jest_studentem = {jest_studentem}, typ: {type(jest_studentem)}")
print(f"ma_dlug = {ma_dlug}, typ: {type(ma_dlug)}")

print("\n--- Bool z por√≥wna≈Ñ ---")
wiek = 20
print(f"wiek = {wiek}")
print(f"wiek > 18: {wiek > 18}")
print(f"wiek == 20: {wiek == 20}")
print(f"wiek != 25: {wiek != 25}")
print(f"wiek < 10: {wiek < 10}")

print("\n--- Bool to podtyp int! ---")
# True = 1, False = 0
print(f"True + True = {True + True}")
print(f"True * 10 = {True * 10}")
print(f"False + 5 = {False + 5}")
print(f"int(True) = {int(True)}")
print(f"int(False) = {int(False)}")

print("\n--- Warto≈õci 'prawdziwe' i 'fa≈Çszywe' (Truthy/Falsy) ---")
# Warto≈õci FA≈ÅSZYWE (Falsy):
print("Warto≈õci Falsy (konwertowane na False):")
falsy_values = [0, 0.0, "", [], {}, set(), None, False]
for val in falsy_values:
    print(f"  bool({str(val):10s}) = {bool(val)}")

print("\nWarto≈õci Truthy (konwertowane na True):")
truthy_values = [1, -1, 3.14, "tekst", [1, 2], {"a": 1}, True]
for val in truthy_values:
    print(f"  bool({str(val):10s}) = {bool(val)}")
```

### Konwersja typ√≥w (rzutowanie)

```python
# ============================================
# KOM√ìRKA 6: Konwersja miƒôdzy typami
# ============================================

print("=== KONWERSJA TYP√ìW ===\n")

print("--- int() - konwersja na liczbƒô ca≈ÇkowitƒÖ ---")
print(f'int("42") = {int("42")}')
print(f'int(3.99) = {int(3.99)}')         # OBCINA, nie zaokrƒÖgla!
print(f'int(True) = {int(True)}')
print(f'int(-3.7) = {int(-3.7)}')          # obcina w stronƒô zera

print("\n--- float() - konwersja na liczbƒô zmiennoprzecinkowƒÖ ---")
print(f'float("3.14") = {float("3.14")}')
print(f'float(42) = {float(42)}')
print(f'float("1e3") = {float("1e3")}')
print(f'float(True) = {float(True)}')

print("\n--- str() - konwersja na tekst ---")
print(f'str(42) = "{str(42)}"')
print(f'str(3.14) = "{str(3.14)}"')
print(f'str(True) = "{str(True)}"')
print(f'str(None) = "{str(None)}"')

print("\n--- bool() - konwersja na warto≈õƒá logicznƒÖ ---")
print(f'bool(1) = {bool(1)}')
print(f'bool(0) = {bool(0)}')
print(f'bool("tekst") = {bool("tekst")}')
print(f'bool("") = {bool("")}')
print(f'bool(None) = {bool(None)}')

print("\n--- ‚ö†Ô∏è Typowe b≈Çƒôdy konwersji ---")
try:
    wynik = int("3.14")  # NIE DZIA≈ÅA! String musi zawieraƒá int
except ValueError as e:
    print(f'int("3.14") ‚Üí B≈ÅƒÑD: {e}')

try:
    wynik = int("abc")   # NIE DZIA≈ÅA!
except ValueError as e:
    print(f'int("abc") ‚Üí B≈ÅƒÑD: {e}')

# RozwiƒÖzanie: najpierw float(), potem int()
print(f'\nint(float("3.14")) = {int(float("3.14"))}')
```

### Sprawdzanie typ√≥w i wieloprzypisanie

```python
# ============================================
# KOM√ìRKA 7: Sprawdzanie typ√≥w i triki
# ============================================

print("=== SPRAWDZANIE TYP√ìW ===\n")

x = 42
print(f"type({x}) = {type(x)}")
print(f"isinstance({x}, int) = {isinstance(x, int)}")
print(f"isinstance({x}, (int, float)) = {isinstance(x, (int, float))}")

print("\n=== WIELOPRZYPISANIE ===\n")

# Przypisanie wielu zmiennych na raz
a, b, c = 1, 2.5, "trzy"
print(f"a={a}, b={b}, c={c}")

# Ta sama warto≈õƒá do wielu zmiennych
x = y = z = 0
print(f"x={x}, y={y}, z={z}")

# Zamiana warto≈õci (swap) - elegancki spos√≥b Pythona!
a, b = 10, 20
print(f"\nPrzed zamianƒÖ: a={a}, b={b}")
a, b = b, a  # W innych jƒôzykach potrzebna jest zmienna tymczasowa
print(f"Po zamianie:  a={a}, b={b}")

print("\n=== DYNAMICZNE TYPOWANIE ===\n")
# Zmienna mo≈ºe zmieniaƒá typ!
x = 42
print(f"x = {x}, typ: {type(x).__name__}")
x = "teraz jestem tekstem"
print(f"x = {x}, typ: {type(x).__name__}")
x = [1, 2, 3]
print(f"x = {x}, typ: {type(x).__name__}")

print("\n=== NONE - brak warto≈õci ===\n")
wynik = None
print(f"wynik = {wynik}, typ: {type(wynik)}")
print(f"wynik is None: {wynik is None}")     # preferowany spos√≥b
print(f"wynik == None: {wynik == None}")       # dzia≈Ça, ale niezalecany
```

## 1.2 Operatory

### Operatory arytmetyczne

```python
# ============================================
# KOM√ìRKA 8: Operatory arytmetyczne
# ============================================

print("=== OPERATORY ARYTMETYCZNE ===\n")

a = 17
b = 5

print(f"a = {a}, b = {b}")
print(f"{'='*40}")

print(f"a + b  = {a + b:>6}    # Dodawanie")
print(f"a - b  = {a - b:>6}    # Odejmowanie")
print(f"a * b  = {a * b:>6}    # Mno≈ºenie")
print(f"a / b  = {a / b:>6}    # Dzielenie (zawsze float!)")
print(f"a // b = {a // b:>6}    # Dzielenie ca≈Çkowite (floor)")
print(f"a % b  = {a % b:>6}    # Reszta z dzielenia (modulo)")
print(f"a ** b = {a ** b:>6}    # Potƒôgowanie")

print(f"\n--- Dzielenie: / vs // ---")
print(f"7 / 2  = {7 / 2}")      # 3.5 (zawsze float)
print(f"7 // 2 = {7 // 2}")      # 3   (zaokrƒÖgla w d√≥≈Ç)
print(f"6 / 2  = {6 / 2}")      # 3.0 (float mimo wyniku ca≈Çkowitego!)
print(f"-7 // 2 = {-7 // 2}")    # -4  (zaokrƒÖgla W D√ì≈Å, nie do zera!)

print(f"\n--- Modulo (%) - zastosowania ---")
# Sprawdzanie parzysto≈õci
for liczba in range(1, 11):
    parzystosc = "parzysta" if liczba % 2 == 0 else "nieparzysta"
    print(f"  {liczba:2d} jest {parzystosc}")

print(f"\n--- Potƒôgowanie ---")
print(f"2 ** 10 = {2 ** 10}")         # 1024
print(f"9 ** 0.5 = {9 ** 0.5}")       # pierwiastek kwadratowy = 3.0
print(f"8 ** (1/3) = {8 ** (1/3)}")   # pierwiastek sze≈õcienny = 2.0

print(f"\n--- Kolejno≈õƒá dzia≈Ça≈Ñ (PEMDAS) ---")
wynik = 2 + 3 * 4 ** 2 - 6 / 3
print(f"2 + 3 * 4 ** 2 - 6 / 3 = {wynik}")
print(f"Kolejno≈õƒá: 4**2=16, 3*16=48, 6/3=2, 2+48-2 = 48")
print(f"\nPriorytety (od najwy≈ºszego):")
print(f"  1. **       potƒôgowanie")
print(f"  2. * / // % mno≈ºenie, dzielenie")
print(f"  3. + -      dodawanie, odejmowanie")
```

### Operatory przypisania

```python
# ============================================
# KOM√ìRKA 9: Operatory przypisania
# ============================================

print("=== OPERATORY PRZYPISANIA ===\n")

# Podstawowy
x = 10
print(f"x = 10       ‚Üí x = {x}")

# Skr√≥cone operatory przypisania
x += 5    # to samo co: x = x + 5
print(f"x += 5       ‚Üí x = {x}")

x -= 3    # to samo co: x = x - 3
print(f"x -= 3       ‚Üí x = {x}")

x *= 2    # to samo co: x = x * 2
print(f"x *= 2       ‚Üí x = {x}")

x /= 4    # to samo co: x = x / 4
print(f"x /= 4       ‚Üí x = {x}")

x = 17
x //= 5   # to samo co: x = x // 5
print(f"\nx = 17")
print(f"x //= 5      ‚Üí x = {x}")

x = 17
x %= 5     # to samo co: x = x % 5
print(f"\nx = 17")
print(f"x %= 5       ‚Üí x = {x}")

x = 2
x **= 10   # to samo co: x = x ** 10
print(f"\nx = 2")
print(f"x **= 10     ‚Üí x = {x}")

print(f"\n‚ö†Ô∏è UWAGA: Python NIE MA operator√≥w ++ i --")
print(f"   Zamiast x++ u≈ºyj: x += 1")
print(f"   Zamiast x-- u≈ºyj: x -= 1")
```

### Operatory por√≥wnania

```python
# ============================================
# KOM√ìRKA 10: Operatory por√≥wnania
# ============================================

print("=== OPERATORY POR√ìWNANIA ===\n")

a = 10
b = 20
c = 10

print(f"a = {a}, b = {b}, c = {c}")
print(f"{'='*45}")

print(f"a == b  ‚Üí {a == b:>5}    # R√≥wne")
print(f"a == c  ‚Üí {a == c:>5}    # R√≥wne")
print(f"a != b  ‚Üí {a != b:>5}    # R√≥≈ºne")
print(f"a > b   ‚Üí {a > b:>5}    # Wiƒôksze")
print(f"a < b   ‚Üí {a < b:>5}    # Mniejsze")
print(f"a >= c  ‚Üí {a >= c:>5}    # Wiƒôksze lub r√≥wne")
print(f"a <= b  ‚Üí {a <= b:>5}    # Mniejsze lub r√≥wne")

print(f"\n--- Por√≥wnania ≈Ça≈Ñcuchowe (unikalne dla Pythona!) ---")
x = 15
print(f"x = {x}")
print(f"10 < x < 20  ‚Üí {10 < x < 20}")    # zamiast: x > 10 and x < 20
print(f"1 < x < 10   ‚Üí {1 < x < 10}")
print(f"10 <= x <= 20 ‚Üí {10 <= x <= 20}")

print(f"\n--- Por√≥wnywanie string√≥w ---")
# Por√≥wnanie leksykograficzne (wg kod√≥w Unicode)
print(f"'abc' == 'abc' ‚Üí {'abc' == 'abc'}")
print(f"'abc' < 'abd'  ‚Üí {'abc' < 'abd'}")
print(f"'abc' < 'ABC'  ‚Üí {'abc' < 'ABC'}")  # ma≈Çe > wielkie w Unicode!
print(f"'A' < 'a'      ‚Üí {'A' < 'a'}")
print(f"ord('A') = {ord('A')}, ord('a') = {ord('a')}")

print(f"\n--- == vs is ---")
# == por√≥wnuje WARTO≈öCI
# is por√≥wnuje TO≈ªSAMO≈öƒÜ (czy to ten sam obiekt w pamiƒôci)
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"a = {a}, b = {b}")
print(f"a == b ‚Üí {a == b}")    # True  - te same warto≈õci
print(f"a is b ‚Üí {a is b}")    # False - r√≥≈ºne obiekty!
print(f"a is c ‚Üí {a is c}")    # True  - ten sam obiekt

print(f"\nid(a) = {id(a)}")
print(f"id(b) = {id(b)}")
print(f"id(c) = {id(c)}")

# WyjƒÖtek: ma≈Çe inty i kr√≥tkie stringi sƒÖ cachowane
x = 5
y = 5
print(f"\nx = y = 5")
print(f"x is y ‚Üí {x is y}")    # True! Python cachuje ma≈Çe liczby (-5 do 256)
```

### Operatory logiczne

```python
# ============================================
# KOM√ìRKA 11: Operatory logiczne
# ============================================

print("=== OPERATORY LOGICZNE ===\n")

print("--- Tabela prawdy AND ---")
print(f"True  and True  = {True and True}")
print(f"True  and False = {True and False}")
print(f"False and True  = {False and True}")
print(f"False and False = {False and False}")

print(f"\n--- Tabela prawdy OR ---")
print(f"True  or True  = {True or True}")
print(f"True  or False = {True or False}")
print(f"False or True  = {False or True}")
print(f"False or False = {False or False}")

print(f"\n--- NOT (negacja) ---")
print(f"not True  = {not True}")
print(f"not False = {not False}")

print(f"\n--- Praktyczne przyk≈Çady ---")
wiek = 25
ma_bilet = True
jest_vip = False

# AND - obie muszƒÖ byƒá prawdziwe
moze_wejsc = wiek >= 18 and ma_bilet
print(f"wiek={wiek}, ma_bilet={ma_bilet}")
print(f"Mo≈ºe wej≈õƒá (>=18 AND bilet)? {moze_wejsc}")

# OR - wystarczy jedna prawdziwa
priorytetowe_wejscie = jest_vip or wiek > 65
print(f"\njest_vip={jest_vip}, wiek={wiek}")
print(f"Priorytetowe wej≈õcie (VIP OR >65)? {priorytetowe_wejscie}")

# NOT - negacja
nie_jest_vip = not jest_vip
print(f"\nnot jest_vip = {nie_jest_vip}")

# Z≈Ço≈ºone wyra≈ºenia
ocena = 85
frekwencja = 90
projekty_oddane = True

zdal = (ocena >= 60 and frekwencja >= 80) or projekty_oddane
print(f"\nocena={ocena}, frekwencja={frekwencja}, projekty={projekty_oddane}")
print(f"Czy zda≈Ç? {zdal}")

print(f"\n--- Short-circuit evaluation (leniwe warto≈õciowanie) ---")
# Python nie sprawdza drugiego warunku, je≈õli pierwszy wystarczy

# AND: je≈õli pierwszy jest False, drugi NIE jest sprawdzany
print("False and print('To siƒô NIE wy≈õwietli')")
result = False and print("To siƒô NIE wy≈õwietli")
print(f"Wynik: {result}")

# OR: je≈õli pierwszy jest True, drugi NIE jest sprawdzany
print("\nTrue or print('To siƒô NIE wy≈õwietli')")
result = True or print("To siƒô NIE wy≈õwietli")
print(f"Wynik: {result}")

print(f"\n--- and/or zwracajƒÖ WARTO≈öCI, nie zawsze bool! ---")
# and: zwraca pierwszy Falsy lub ostatni element
print(f"'hello' and 'world' = {'hello' and 'world'}")  # 'world'
print(f"'' and 'world'      = {repr('' and 'world')}")       # ''
print(f"0 and 42             = {0 and 42}")                    # 0

# or: zwraca pierwszy Truthy lub ostatni element
print(f"\n'hello' or 'world' = {'hello' or 'world'}")  # 'hello'
print(f"'' or 'world'      = {'' or 'world'}")          # 'world'
print(f"0 or 42             = {0 or 42}")                # 42
print(f"'' or 0 or None     = {repr('' or 0 or None)}")  # None

# Praktyczne u≈ºycie: warto≈õƒá domy≈õlna
nazwa_uzytkownika = ""
wyswietlana_nazwa = nazwa_uzytkownika or "Anonim"
print(f"\nnazwa_uzytkownika = '{nazwa_uzytkownika}'")
print(f"Wy≈õwietlana nazwa: '{wyswietlana_nazwa}'")
```

### F-strings ‚Äì wyr√≥wnywanie i padding

```python
# ============================================
# KOM√ìRKA 16: F-strings - wyr√≥wnywanie
# ============================================

print("=== F-STRINGS - WYR√ìWNYWANIE ===\n")

# Sk≈Çadnia: {warto≈õƒá:wype≈Çniacz wyr√≥wnanie szeroko≈õƒá}
#   < = wyr√≥wnaj do lewej
#   > = wyr√≥wnaj do prawej
#   ^ = wy≈õrodkuj

print("--- Wyr√≥wnywanie tekstu ---")
tekst = "Python"
print(f"|{tekst:<20}| ‚Üê lewo (domy≈õlne dla string√≥w)")
print(f"|{tekst:>20}| ‚Üê prawo")
print(f"|{tekst:^20}| ‚Üê ≈õrodek")

print(f"\n--- Wyr√≥wnywanie z wype≈Çniaczem ---")
print(f"|{tekst:.<20}| ‚Üê kropki, lewo")
print(f"|{tekst:.>20}| ‚Üê kropki, prawo")
print(f"|{tekst:*^20}| ‚Üê gwiazdki, ≈õrodek")
print(f"|{tekst:-^20}| ‚Üê my≈õlniki, ≈õrodek")
print(f"|{tekst:=^20}| ‚Üê znaki r√≥wno≈õci, ≈õrodek")

print(f"\n--- Wyr√≥wnywanie liczb ---")
for i in [1, 42, 100, 9999, 123456]:
    print(f"|{i:>10,}|")

print(f"\n--- Wype≈Çnianie zerami ---")
for i in [1, 42, 100, 9999]:
    print(f"Numer: {i:05d}")

print(f"\n--- Praktyczny przyk≈Çad: tabelka ---")
print(f"{'Produkt':<15} {'Ilo≈õƒá':>6} {'Cena':>10} {'Suma':>10}")
print(f"{'-'*15} {'-'*6} {'-'*10} {'-'*10}")

produkty = [
    ("Jab≈Çka", 5, 3.50),
    ("Chleb", 2, 4.99),
    ("Mleko", 3, 5.20),
    ("Mas≈Ço", 1, 7.49),
]

total = 0
for nazwa, ilosc, cena in produkty:
    suma = ilosc * cena
    total += suma
    print(f"{nazwa:<15} {ilosc:>6} {cena:>10.2f} {suma:>10.2f}")

print(f"{'-'*15} {'-'*6} {'-'*10} {'-'*10}")
print(f"{'RAZEM':<15} {'':>6} {'':>10} {total:>10.2f}")
```

### Docstrings ‚Äì dokumentowanie kodu

```python
# ============================================
# KOM√ìRKA 20: Docstrings - podstawy
# ============================================

print("=== DOCSTRINGS ===\n")

# Docstring to specjalny string dokumentujƒÖcy modu≈Ç, klasƒô lub funkcjƒô
# Umieszczany ZARAZ PO definicji (jako pierwszy element)
# Dostƒôpny przez atrybut __doc__ lub funkcjƒô help()


# --- Docstring dla funkcji ---

def powitaj(imie):
    """Wy≈õwietla powitanie dla podanej osoby."""
    print(f"Cze≈õƒá, {imie}!")


def oblicz_bmi(waga, wzrost):
    """
    Oblicza wska≈∫nik masy cia≈Ça (BMI).

    Parametry:
        waga (float): Waga w kilogramach
        wzrost (float): Wzrost w metrach

    Zwraca:
        float: Warto≈õƒá BMI zaokrƒÖglona do 1 miejsca

    Przyk≈Çad:
        >>> oblicz_bmi(70, 1.75)
        22.9
    """
    bmi = waga / (wzrost ** 2)
    return round(bmi, 1)


# U≈ºycie funkcji
powitaj("Anna")
wynik = oblicz_bmi(70, 1.75)
print(f"BMI: {wynik}")

# Dostƒôp do docstringa
print(f"\n--- Dostƒôp do docstringa ---")
print(f"powitaj.__doc__: {powitaj.__doc__}")
print(f"\noblicz_bmi.__doc__:")
print(oblicz_bmi.__doc__)
```

### Docstrings dla klas i modu≈Ç√≥w

```python
# ============================================
# KOM√ìRKA 22: Docstrings dla klas i modu≈Ç√≥w
# ============================================

print("=== DOCSTRINGS DLA KLAS ===\n")


class KontoBankowe:
    """Reprezentuje konto bankowe klienta.

    Klasa umo≈ºliwia podstawowe operacje bankowe:
    wp≈Çaty, wyp≈Çaty i sprawdzanie salda.

    Attributes:
        wlasciciel (str): Imiƒô i nazwisko w≈Ça≈õciciela konta.
        saldo (float): Aktualne saldo konta.
        historia (list): Lista wykonanych transakcji.

    Example:
        >>> konto = KontoBankowe("Jan Kowalski", 1000)
        >>> konto.wplata(500)
        >>> print(konto.saldo)
        1500
    """

    def __init__(self, wlasciciel, saldo_poczatkowe=0):
        """Inicjalizuje nowe konto bankowe.

        Args:
            wlasciciel (str): Imiƒô i nazwisko w≈Ça≈õciciela.
            saldo_poczatkowe (float, optional): PoczƒÖtkowe saldo.
                Domy≈õlnie 0.
        """
        self.wlasciciel = wlasciciel
        self.saldo = saldo_poczatkowe
        self.historia = []

    def wplata(self, kwota):
        """Dokonuje wp≈Çaty na konto.

        Args:
            kwota (float): Kwota do wp≈Çaty (musi byƒá dodatnia).

        Raises:
            ValueError: Je≈õli kwota jest ujemna lub zerowa.
        """
        if kwota <= 0:
            raise ValueError("Kwota wp≈Çaty musi byƒá dodatnia!")
        self.saldo += kwota
        self.historia.append(f"Wp≈Çata: +{kwota:.2f}")

    def wyplata(self, kwota):
        """Dokonuje wyp≈Çaty z konta.

        Args:
            kwota (float): Kwota do wyp≈Çaty.

        Returns:
            float: Wyp≈Çacona kwota.

        Raises:
            ValueError: Je≈õli kwota jest ujemna.
            ValueError: Je≈õli brak wystarczajƒÖcych ≈õrodk√≥w.
        """
        if kwota <= 0:
            raise ValueError("Kwota wyp≈Çaty musi byƒá dodatnia!")
        if kwota > self.saldo:
            raise ValueError(f"Brak ≈õrodk√≥w! Saldo: {self.saldo:.2f}")
        self.saldo -= kwota
        self.historia.append(f"Wyp≈Çata: -{kwota:.2f}")
        return kwota

    def __str__(self):
        """Zwraca tekstowƒÖ reprezentacjƒô konta."""
        return f"Konto({self.wlasciciel}, saldo={self.saldo:.2f})"


# U≈ºycie klasy
konto = KontoBankowe("Anna Nowak", 1000)
konto.wplata(500)
konto.wyplata(200)
print(konto)
print(f"Historia: {konto.historia}")

# Dostƒôp do docstring√≥w
print(f"\n--- Docstring klasy ---")
print(KontoBankowe.__doc__)

print(f"\n--- Docstring metody ---")
print(KontoBankowe.wplata.__doc__)

# help() w JupyterLab - odkomentuj poni≈ºsze
# help(KontoBankowe)
```

### ƒÜwiczenia do samodzielnego wykonania

```python
# ============================================
# KOM√ìRKA 24: ƒÜwiczenia
# ============================================

print("=== ƒÜWICZENIA DO SAMODZIELNEGO WYKONANIA ===\n")

# -----------------------------------------------
# ƒÜWICZENIE 1: Kalkulator BMI
# Stw√≥rz zmienne waga i wzrost, oblicz BMI,
# wy≈õwietl wynik sformatowany do 1 miejsca po
# przecinku z opisem kategorii
# -----------------------------------------------

# Tw√≥j kod tutaj:
waga = 75       # kg
wzrost = 1.80   # m

# BMI = waga / wzrost¬≤
# < 18.5 = niedowaga
# 18.5-24.9 = norma
# 25-29.9 = nadwaga
# >= 30 = oty≈Ço≈õƒá

bmi = waga / (wzrost ** 2)
if bmi < 18.5:
    kategoria = "niedowaga"
elif bmi < 25:
    kategoria = "norma"
elif bmi < 30:
    kategoria = "nadwaga"
else:
    kategoria = "oty≈Ço≈õƒá"

print(f"Waga: {waga} kg, Wzrost: {wzrost} m")
print(f"BMI: {bmi:.1f} ({kategoria})")


# -----------------------------------------------
# ƒÜWICZENIE 2: Paragon ze sklepu
# Stw√≥rz zmienne z nazwami produkt√≥w, cenami
# i ilo≈õciami, wy≈õwietl sformatowany paragon
# -----------------------------------------------

print(f"\n{'='*40}")
print(f"{'PARAGON':^40}")
print(f"{'='*40}")

zakupy = [
    ("Chleb ≈ºytni", 2, 4.50),
    ("Mas≈Ço extra", 1, 7.99),
    ("Mleko 3.2%", 3, 4.29),
    ("Ser ≈º√≥≈Çty", 1, 12.99),
]

suma = 0
for nazwa, ilosc, cena in zakupy:
    wartosc = ilosc * cena
    suma += wartosc
    print(f"{nazwa:<15} {ilosc}x {cena:>6.2f} = {wartosc:>8.2f}")

print(f"{'-'*40}")
print(f"{'SUMA:':<25} {suma:>13.2f} PLN")
print(f"{'='*40}")


# -----------------------------------------------
# ƒÜWICZENIE 3: Sprawd≈∫, czy liczba jest podzielna
# przez 3 I 5 jednocze≈õnie (FizzBuzz preview)
# -----------------------------------------------

print(f"\n--- ƒÜwiczenie 3: Podzielno≈õƒá ---")
for liczba in range(1, 21):
    if liczba % 3 == 0 and liczba % 5 == 0:
        wynik = "FizzBuzz"
    elif liczba % 3 == 0:
        wynik = "Fizz"
    elif liczba % 5 == 0:
        wynik = "Buzz"
    else:
        wynik = str(liczba)
    print(f"{liczba:>2}: {wynik}")
```
