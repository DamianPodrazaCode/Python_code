# 2. Zmienne - Fundamenty

### 2.1 Czym jest zmienna?

```python
# Zmienna to NAZWA wskazująca na OBIEKT w pamięci

#    Nazwa          Obiekt
#    ┌───┐         ┌─────┐
#    │ x │ ──────> │  5  │
#    └───┘         └─────┘

x = 5

# W Pythonie zmienne są REFERENCJAMI (wskaźnikami) do obiektów
# NIE są "pudełkami" przechowującymi wartości

# Sprawdź adres obiektu w pamięci:
print(id(x))  # np. 140234567890
print(hex(id(x)))  # np. 0x7f8a12345678
```

### 2.2 Tworzenie zmiennych

```python
# Zmienna powstaje w momencie pierwszego przypisania
nazwa = "wartość"

# Nie trzeba deklarować typu (dynamiczne typowanie)
liczba = 42          # int
pi = 3.14159         # float
imie = "Anna"        # str
aktywny = True       # bool

# Można przypisać wiele zmiennych naraz
a, b, c = 1, 2, 3
print(a, b, c)  # 1 2 3

# Ta sama wartość do wielu zmiennych
x = y = z = 0
print(x, y, z)  # 0 0 0

# Zamiana wartości (swap) - elegancki Python
a, b = 1, 2
a, b = b, a  # Zamiana bez zmiennej tymczasowej!
print(a, b)  # 2 1

# Rozpakowywanie (unpacking)
punkt = (10, 20, 30)
x, y, z = punkt
print(x, y, z)  # 10 20 30

# Rozpakowywanie z gwiazdką
pierwszy, *środek, ostatni = [1, 2, 3, 4, 5]
print(pierwszy)  # 1
print(środek)    # [2, 3, 4]
print(ostatni)   # 5
```

# 3. Typy Danych Liczbowych

### 3.1 Integer (int) - Liczby całkowite

```python
# Podstawowe tworzenie
x = 42
y = -17
z = 0

# Python 3 ma nieograniczoną precyzję dla int!
wielka_liczba = 123456789012345678901234567890
print(wielka_liczba * 2)  # Działa bez problemu!

# Różne systemy liczbowe
dziesietny = 255
binarny = 0b11111111      # prefiks 0b
osemkowy = 0o377          # prefiks 0o
szesnastkowy = 0xFF       # prefiks 0x lub 0X

print(dziesietny == binarny == osemkowy == szesnastkowy)  # True

# Konwersje między systemami
print(bin(255))   # '0b11111111'
print(oct(255))   # '0o377'
print(hex(255))   # '0xff'

# Parsowanie stringów
print(int('255'))       # 255 (dziesiętny)
print(int('11111111', 2))  # 255 (z binarnego)
print(int('FF', 16))    # 255 (z szesnastkowego)

# Separatory dla czytelności (Python 3.6+)
milion = 1_000_000
miliard = 1_000_000_000
binarny = 0b1111_1111_0000_0000
szesnastkowy = 0xaabb_ccdd

# Operacje na bitach - przydatne w niskopoziomowym programowaniu
print(255 .bit_length())  # 8 - liczba bitów potrzebnych do reprezentacji

# Sprawdzanie typów
print(type(42))       # <class 'int'>
print(type(True))     # <class 'bool'> - bool to podtyp int!
print(isinstance(True, int))  # True

# Ciekawe fakty o bool
print(True + True)   # 2
print(True * 10)     # 10
print(False + 5)     # 5
```

### 3.2 Float - Liczby zmiennoprzecinkowe

```python
# Podstawowe tworzenie
x = 3.14
y = -0.001
z = 2.0  # to jest float, nie int

# Notacja naukowa (scientific notation)
avogadro = 6.022e23      # 6.022 × 10²³
planck = 6.626e-34       # 6.626 × 10⁻³⁴
mega = 1e6               # 1,000,000.0

print(avogadro)  # 6.022e+23
print(planck)    # 6.626e-34

# Float jest przechowywany jako IEEE 754 double (64-bit)
# Ograniczona precyzja!

# ⚠️ UWAGA: Problemy z precyzją float
print(0.1 + 0.2)  # 0.30000000000000004 (nie 0.3!)
print(0.1 + 0.2 == 0.3)  # False!

# Dlaczego? 0.1 nie da się dokładnie przedstawić w systemie binarnym
print(f"{0.1:.20f}")  # 0.10000000000000000555

# Rozwiązania problemu precyzji:

# 1. Porównywanie z tolerancją
def almost_equal(a, b, tolerance=1e-9):
    return abs(a - b) < tolerance

print(almost_equal(0.1 + 0.2, 0.3))  # True

# 2. Użycie math.isclose()
import math
print(math.isclose(0.1 + 0.2, 0.3))  # True

# 3. Dla obliczeń finansowych - moduł decimal
from decimal import Decimal
print(Decimal('0.1') + Decimal('0.2') == Decimal('0.3'))  # True

# 4. Dla ułamków - moduł fractions
from fractions import Fraction
print(Fraction(1, 10) + Fraction(2, 10) == Fraction(3, 10))  # True

# Specjalne wartości float
print(float('inf'))   # inf (nieskończoność)
print(float('-inf'))  # -inf
print(float('nan'))   # nan (Not a Number)

# Sprawdzanie specjalnych wartości
print(math.isinf(float('inf')))   # True
print(math.isnan(float('nan')))   # True

# Właściwości float
import sys
print(sys.float_info.max)      # ~1.8e+308 (maksimum)
print(sys.float_info.min)      # ~2.2e-308 (minimum dodatnie)
print(sys.float_info.epsilon)  # ~2.2e-16 (najmniejsza różnica)

# Metody float
x = 3.7
print(x.is_integer())     # False
print((3.0).is_integer()) # True
print(x.as_integer_ratio())  # (8316386860379603, 2251799813685248)
```

### 3.3 Complex - Liczby zespolone

```python
# Tworzenie liczb zespolonych
z = 3 + 4j        # j oznacza jednostkę urojoną
z2 = complex(3, 4)  # alternatywna składnia

print(z)   # (3+4j)
print(z2)  # (3+4j)

# Dostęp do części rzeczywistej i urojonej
print(z.real)  # 3.0
print(z.imag)  # 4.0

# Sprzężenie zespolone
print(z.conjugate())  # (3-4j)

# Moduł (wartość bezwzględna)
print(abs(z))  # 5.0 (√(3² + 4²))

# Operacje
z1 = 1 + 2j
z2 = 3 + 4j
print(z1 + z2)  # (4+6j)
print(z1 * z2)  # (-5+10j)
print(z1 / z2)  # (0.44+0.08j)

# W ML/PyTorch - tensory zespolone
import torch
t = torch.tensor([1+2j, 3+4j])
print(t.dtype)  # torch.complex64
```
