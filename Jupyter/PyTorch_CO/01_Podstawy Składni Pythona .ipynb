# 3. Typy Danych Liczbowych

### 3.1 Integer (int) - Liczby ca≈Çkowite

```python
# Podstawowe tworzenie
x = 42
y = -17
z = 0

# Python 3 ma nieograniczonƒÖ precyzjƒô dla int!
wielka_liczba = 123456789012345678901234567890
print(wielka_liczba * 2)  # Dzia≈Ça bez problemu!

# R√≥≈ºne systemy liczbowe
dziesietny = 255
binarny = 0b11111111      # prefiks 0b
osemkowy = 0o377          # prefiks 0o
szesnastkowy = 0xFF       # prefiks 0x lub 0X

print(dziesietny == binarny == osemkowy == szesnastkowy)  # True

# Konwersje miƒôdzy systemami
print(bin(255))   # '0b11111111'
print(oct(255))   # '0o377'
print(hex(255))   # '0xff'

# Parsowanie string√≥w
print(int('255'))       # 255 (dziesiƒôtny)
print(int('11111111', 2))  # 255 (z binarnego)
print(int('FF', 16))    # 255 (z szesnastkowego)

# Separatory dla czytelno≈õci (Python 3.6+)
milion = 1_000_000
miliard = 1_000_000_000
binarny = 0b1111_1111_0000_0000
szesnastkowy = 0xaabb_ccdd

# Operacje na bitach - przydatne w niskopoziomowym programowaniu
print(255 .bit_length())  # 8 - liczba bit√≥w potrzebnych do reprezentacji

# Sprawdzanie typ√≥w
print(type(42))       # <class 'int'>
print(type(True))     # <class 'bool'> - bool to podtyp int!
print(isinstance(True, int))  # True

# Ciekawe fakty o bool
print(True + True)   # 2
print(True * 10)     # 10
print(False + 5)     # 5
```

### 3.2 Float - Liczby zmiennoprzecinkowe

```python
# Podstawowe tworzenie
x = 3.14
y = -0.001
z = 2.0  # to jest float, nie int

# Notacja naukowa (scientific notation)
avogadro = 6.022e23      # 6.022 √ó 10¬≤¬≥
planck = 6.626e-34       # 6.626 √ó 10‚Åª¬≥‚Å¥
mega = 1e6               # 1,000,000.0

print(avogadro)  # 6.022e+23
print(planck)    # 6.626e-34

# Float jest przechowywany jako IEEE 754 double (64-bit)
# Ograniczona precyzja!

# ‚ö†Ô∏è UWAGA: Problemy z precyzjƒÖ float
print(0.1 + 0.2)  # 0.30000000000000004 (nie 0.3!)
print(0.1 + 0.2 == 0.3)  # False!

# Dlaczego? 0.1 nie da siƒô dok≈Çadnie przedstawiƒá w systemie binarnym
print(f"{0.1:.20f}")  # 0.10000000000000000555

# RozwiƒÖzania problemu precyzji:

# 1. Por√≥wnywanie z tolerancjƒÖ
def almost_equal(a, b, tolerance=1e-9):
    return abs(a - b) < tolerance

print(almost_equal(0.1 + 0.2, 0.3))  # True

# 2. U≈ºycie math.isclose()
import math
print(math.isclose(0.1 + 0.2, 0.3))  # True

# 3. Dla oblicze≈Ñ finansowych - modu≈Ç decimal
from decimal import Decimal
print(Decimal('0.1') + Decimal('0.2') == Decimal('0.3'))  # True

# 4. Dla u≈Çamk√≥w - modu≈Ç fractions
from fractions import Fraction
print(Fraction(1, 10) + Fraction(2, 10) == Fraction(3, 10))  # True

# Specjalne warto≈õci float
print(float('inf'))   # inf (niesko≈Ñczono≈õƒá)
print(float('-inf'))  # -inf
print(float('nan'))   # nan (Not a Number)

# Sprawdzanie specjalnych warto≈õci
print(math.isinf(float('inf')))   # True
print(math.isnan(float('nan')))   # True

# W≈Ça≈õciwo≈õci float
import sys
print(sys.float_info.max)      # ~1.8e+308 (maksimum)
print(sys.float_info.min)      # ~2.2e-308 (minimum dodatnie)
print(sys.float_info.epsilon)  # ~2.2e-16 (najmniejsza r√≥≈ºnica)

# Metody float
x = 3.7
print(x.is_integer())     # False
print((3.0).is_integer()) # True
print(x.as_integer_ratio())  # (8316386860379603, 2251799813685248)
```

### 3.3 Complex - Liczby zespolone

```python
# Tworzenie liczb zespolonych
z = 3 + 4j        # j oznacza jednostkƒô urojonƒÖ
z2 = complex(3, 4)  # alternatywna sk≈Çadnia

print(z)   # (3+4j)
print(z2)  # (3+4j)

# Dostƒôp do czƒô≈õci rzeczywistej i urojonej
print(z.real)  # 3.0
print(z.imag)  # 4.0

# Sprzƒô≈ºenie zespolone
print(z.conjugate())  # (3-4j)

# Modu≈Ç (warto≈õƒá bezwzglƒôdna)
print(abs(z))  # 5.0 (‚àö(3¬≤ + 4¬≤))

# Operacje
z1 = 1 + 2j
z2 = 3 + 4j
print(z1 + z2)  # (4+6j)
print(z1 * z2)  # (-5+10j)
print(z1 / z2)  # (0.44+0.08j)

# W ML/PyTorch - tensory zespolone
import torch
t = torch.tensor([1+2j, 3+4j])
print(t.dtype)  # torch.complex64
```

### 3.4 Hierarchia typ√≥w liczbowych

```python
# Python ma hierarchiƒô typ√≥w liczbowych:
# numbers.Number
# ‚îú‚îÄ‚îÄ numbers.Complex
# ‚îÇ   ‚îî‚îÄ‚îÄ numbers.Real
# ‚îÇ       ‚îî‚îÄ‚îÄ numbers.Rational
# ‚îÇ           ‚îî‚îÄ‚îÄ numbers.Integral

from numbers import Number, Complex, Real, Rational, Integral

# Sprawdzanie
print(isinstance(5, Integral))    # True
print(isinstance(5, Real))        # True
print(isinstance(5, Complex))     # True
print(isinstance(5.0, Integral))  # False
print(isinstance(5.0, Real))      # True
print(isinstance(5+0j, Complex))  # True
print(isinstance(5+0j, Real))     # False

# Mieszane operacje - automatyczna konwersja "w g√≥rƒô"
print(type(5 + 3.14))     # <class 'float'>
print(type(5 + 3.14 + 0j))  # <class 'complex'>
```

### 4.4 Indeksowanie i slicing

```python
s = "Python"
#    P  y  t  h  o  n
#    0  1  2  3  4  5   <- indeksy od poczƒÖtku
#   -6 -5 -4 -3 -2 -1   <- indeksy od ko≈Ñca

# Indeksowanie - pojedynczy znak
print(s[0])    # 'P' - pierwszy
print(s[1])    # 'y' - drugi
print(s[-1])   # 'n' - ostatni
print(s[-2])   # 'o' - przedostatni

# Slicing - zakres [start:stop:step]
print(s[0:3])   # 'Pyt' - znaki 0, 1, 2 (stop NIE jest w≈ÇƒÖczony)
print(s[2:5])   # 'tho'
print(s[:3])    # 'Pyt' - od poczƒÖtku do 3 (bez 3)
print(s[3:])    # 'hon' - od 3 do ko≈Ñca
print(s[:])     # 'Python' - kopia ca≈Çego stringa
print(s[::2])   # 'Pto' - co drugi znak
print(s[1::2])  # 'yhn' - co drugi, zaczynajƒÖc od 1
print(s[::-1])  # 'nohtyP' - odwr√≥cony string!
print(s[-3:])   # 'hon' - ostatnie 3 znaki
print(s[:-3])   # 'Pyt' - wszystko opr√≥cz ostatnich 3

# Slicing zawsze zwraca nowy string
fragment = s[1:4]
print(type(fragment))  # <class 'str'>
```

### 4.6 Metody string√≥w

```python
s = "  Hello, World!  "

# === WIELKO≈öƒÜ LITER ===
print(s.upper())       # "  HELLO, WORLD!  "
print(s.lower())       # "  hello, world!  "
print(s.capitalize())  # "  hello, world!  " (tylko pierwsza litera stringa)
print(s.title())       # "  Hello, World!  " (ka≈ºde s≈Çowo)
print(s.swapcase())    # "  hELLO, wORLD!  "
print("hello".islower())    # True
print("HELLO".isupper())    # True
print("Hello World".istitle())  # True

# === USUWANIE BIA≈ÅYCH ZNAK√ìW ===
print(s.strip())       # "Hello, World!" (z obu stron)
print(s.lstrip())      # "Hello, World!  " (z lewej)
print(s.rstrip())      # "  Hello, World!" (z prawej)
print("xxxHelloxxx".strip('x'))  # "Hello" (usuwa podane znaki)

# === WYSZUKIWANIE ===
s = "Hello, World!"
print(s.find("o"))       # 4 (pierwszy indeks)
print(s.find("o", 5))    # 8 (szukaj od indeksu 5)
print(s.rfind("o"))      # 8 (od ko≈Ñca)
print(s.find("x"))       # -1 (nie znaleziono)
print(s.index("o"))      # 4 (jak find, ale ValueError je≈õli nie ma)
print(s.count("o"))      # 2 (liczba wystƒÖpie≈Ñ)
print(s.startswith("Hello"))  # True
print(s.endswith("!"))        # True

# === SPRAWDZANIE ZAWARTO≈öCI ===
print("123".isdigit())     # True
print("abc".isalpha())     # True
print("abc123".isalnum())  # True (litery lub cyfry)
print("   ".isspace())     # True
print("123".isnumeric())   # True
print("¬Ω".isnumeric())     # True (rozszerzone)
print("123".isdecimal())   # True

# === ZAMIANA ===
s = "Hello, World!"
print(s.replace("World", "Python"))  # "Hello, Python!"
print(s.replace("o", "0", 1))        # "Hell0, World!" (tylko pierwsze)

# === DZIELENIE I ≈ÅƒÑCZENIE ===
s = "apple,banana,cherry"
print(s.split(","))        # ['apple', 'banana', 'cherry']
print(s.split(",", 1))     # ['apple', 'banana,cherry'] (max 1 podzia≈Ç)

s = "Line 1\nLine 2\nLine 3"
print(s.splitlines())      # ['Line 1', 'Line 2', 'Line 3']

words = ['apple', 'banana', 'cherry']
print(", ".join(words))    # "apple, banana, cherry"
print("-".join(words))     # "apple-banana-cherry"

# === WYR√ìWNYWANIE ===
s = "Hello"
print(s.center(11))        # "   Hello   "
print(s.center(11, '-'))   # "---Hello---"
print(s.ljust(10))         # "Hello     "
print(s.rjust(10))         # "     Hello"
print(s.zfill(10))         # "00000Hello"
print("42".zfill(5))       # "00042"

# === KODOWANIE ===
s = "Za≈º√≥≈Çƒá gƒô≈õlƒÖ ja≈∫≈Ñ"
encoded = s.encode('utf-8')  # bytes
print(encoded)  # b'Za\xc5\xbc\xc3\xb3\xc5\x82\xc4\x87...'
decoded = encoded.decode('utf-8')  # z powrotem string
print(decoded)  # "Za≈º√≥≈Çƒá gƒô≈õlƒÖ ja≈∫≈Ñ"
```

# 5. Typ Logiczny - Boolean

### 5.1 Podstawy

```python
# Dwie warto≈õci logiczne
prawda = True
falsz = False

print(type(True))   # <class 'bool'>
print(type(False))  # <class 'bool'>

# Bool jest podtypem int!
print(isinstance(True, int))  # True
print(True == 1)   # True
print(False == 0)  # True
print(True + True)  # 2
print(True * 5)     # 5
```

### 5.2 Operacje logiczne

```python
# AND - oba muszƒÖ byƒá True
print(True and True)    # True
print(True and False)   # False
print(False and True)   # False
print(False and False)  # False

# OR - przynajmniej jeden True
print(True or True)    # True
print(True or False)   # True
print(False or True)   # True
print(False or False)  # False

# NOT - negacja
print(not True)   # False
print(not False)  # True

# Kombinacje
print(True and not False)    # True
print(not (True and False))  # True
print((True or False) and True)  # True
```
