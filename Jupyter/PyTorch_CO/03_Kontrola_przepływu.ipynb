# 3. Kontrola przepływu w Pythonie

## 3.1 Instrukcje warunkowe (if / elif / else)

### Podstawowa składnia

```python
# === PODSTAWOWA INSTRUKCJA IF ===

wiek = 20

if wiek >= 18:
    print("Jesteś pełnoletni/a")
```

```python
# === IF / ELSE ===

wiek = 15

if wiek >= 18:
    print("Jesteś pełnoletni/a")
else:
    print("Jesteś niepełnoletni/a")
```

```python
# === IF / ELIF / ELSE ===
# elif = "else if" - pozwala sprawdzać wiele warunków po kolei

temperatura = 25

if temperatura < 0:
    print("Mróz! Ubierz się ciepło!")
elif temperatura < 10:
    print("Zimno, weź kurtkę")
elif temperatura < 20:
    print("Chłodno, weź bluzę")
elif temperatura < 30:
    print("Przyjemnie ciepło")
else:
    print("Upał! Pij dużo wody!")

print(f"Aktualna temperatura: {temperatura}°C")
```

### Operatory porównania

```python
# === OPERATORY PORÓWNANIA ===

a = 10
b = 20

print("=== Operatory porównania ===")
print(f"a = {a}, b = {b}")
print(f"a == b (równe):          {a == b}")    # False
print(f"a != b (różne):          {a != b}")    # True
print(f"a > b  (większe):        {a > b}")     # False
print(f"a < b  (mniejsze):       {a < b}")     # True
print(f"a >= b (większe-równe):  {a >= b}")    # False
print(f"a <= b (mniejsze-równe): {a <= b}")    # True
```

### Operatory logiczne (and, or, not)

```python
# === OPERATORY LOGICZNE ===

wiek = 25
ma_prawo_jazdy = True
jest_trzezwy = True

# AND - oba warunki muszą być prawdziwe
if wiek >= 18 and ma_prawo_jazdy:
    print("Może prowadzić samochód")

# OR - wystarczy jeden warunek prawdziwy
pogoda = "deszcz"
if pogoda == "deszcz" or pogoda == "śnieg":
    print("Weź parasol!")

# NOT - negacja warunku
zalogowany = False
if not zalogowany:
    print("Proszę się zalogować")

# Łączenie operatorów
print("\n=== Łączenie operatorów ===")
if wiek >= 18 and ma_prawo_jazdy and jest_trzezwy:
    print("Spełnia WSZYSTKIE warunki do prowadzenia")
else:
    print("NIE może prowadzić")
```

### Zagnieżdżone warunki

```python
# === ZAGNIEŻDŻONE WARUNKI ===

wiek = 25
bilet = "ulgowy"
jest_studentem = True

print("=== System biletowy ===")

if wiek < 4:
    print("Przejazd darmowy (dziecko do 4 lat)")
else:
    if bilet == "ulgowy":
        if jest_studentem:
            cena = 1.50
            print(f"Bilet ulgowy studencki: {cena} zł")
        else:
            cena = 2.00
            print(f"Bilet ulgowy: {cena} zł")
    elif bilet == "normalny":
        cena = 3.50
        print(f"Bilet normalny: {cena} zł")
    else:
        print("Nieznany typ biletu!")
```

### Wyrażenie warunkowe (ternary operator)

```python
# === WYRAŻENIE WARUNKOWE (TERNARY) ===
# składnia: wartość_jeśli_tak if warunek else wartość_jeśli_nie

wiek = 20

# Zamiast:
# if wiek >= 18:
#     status = "dorosły"
# else:
#     status = "niepełnoletni"

# Można napisać w jednej linii:
status = "dorosły" if wiek >= 18 else "niepełnoletni"
print(f"Wiek: {wiek}, Status: {status}")

# Więcej przykładów
liczba = -5
typ = "dodatnia" if liczba > 0 else "ujemna" if liczba < 0 else "zero"
print(f"Liczba {liczba} jest {typ}")

# W praktyce - np. formatowanie tekstu
punkty = 1
tekst = f"Masz {punkty} punkt{'y' if punkty > 1 else ''}"
print(tekst)
```

### Sprawdzanie przynależności (in) i tożsamości (is)

```python
# === OPERATOR 'in' ===

# Sprawdzanie czy element jest w kolekcji
owoce = ["jabłko", "banan", "gruszka", "pomarańcza"]

szukany = "banan"
if szukany in owoce:
    print(f"'{szukany}' jest na liście owoców")

# Sprawdzanie w stringu
zdanie = "Python jest świetnym językiem programowania"
if "Python" in zdanie:
    print("Zdanie zawiera słowo 'Python'")

# Sprawdzanie w słowniku (sprawdza klucze!)
oceny = {"matematyka": 5, "fizyka": 4, "chemia": 3}
if "matematyka" in oceny:
    print(f"Ocena z matematyki: {oceny['matematyka']}")

# === OPERATOR 'is' ===
# Sprawdza tożsamość (czy to ten sam obiekt w pamięci)
# NIE jest tym samym co == (porównanie wartości)

a = None
if a is None:
    print("Zmienna 'a' jest None")

# is vs ==
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(f"\nx == y: {x == y}")  # True (te same wartości)
print(f"x is y: {x is y}")    # False (różne obiekty!)
print(f"x is z: {x is z}")    # True (ten sam obiekt)
```
