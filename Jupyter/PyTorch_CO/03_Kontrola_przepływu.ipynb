# 3. Kontrola przepływu w Pythonie

## 3.1 Instrukcje warunkowe (if / elif / else)

### Podstawowa składnia

```python
# === PODSTAWOWA INSTRUKCJA IF ===

wiek = 20

if wiek >= 18:
    print("Jesteś pełnoletni/a")
```

```python
# === IF / ELSE ===

wiek = 15

if wiek >= 18:
    print("Jesteś pełnoletni/a")
else:
    print("Jesteś niepełnoletni/a")
```

```python
# === IF / ELIF / ELSE ===
# elif = "else if" - pozwala sprawdzać wiele warunków po kolei

temperatura = 25

if temperatura < 0:
    print("Mróz! Ubierz się ciepło!")
elif temperatura < 10:
    print("Zimno, weź kurtkę")
elif temperatura < 20:
    print("Chłodno, weź bluzę")
elif temperatura < 30:
    print("Przyjemnie ciepło")
else:
    print("Upał! Pij dużo wody!")

print(f"Aktualna temperatura: {temperatura}°C")
```

### Operatory porównania

```python
# === OPERATORY PORÓWNANIA ===

a = 10
b = 20

print("=== Operatory porównania ===")
print(f"a = {a}, b = {b}")
print(f"a == b (równe):          {a == b}")    # False
print(f"a != b (różne):          {a != b}")    # True
print(f"a > b  (większe):        {a > b}")     # False
print(f"a < b  (mniejsze):       {a < b}")     # True
print(f"a >= b (większe-równe):  {a >= b}")    # False
print(f"a <= b (mniejsze-równe): {a <= b}")    # True
```

### Operatory logiczne (and, or, not)

```python
# === OPERATORY LOGICZNE ===

wiek = 25
ma_prawo_jazdy = True
jest_trzezwy = True

# AND - oba warunki muszą być prawdziwe
if wiek >= 18 and ma_prawo_jazdy:
    print("Może prowadzić samochód")

# OR - wystarczy jeden warunek prawdziwy
pogoda = "deszcz"
if pogoda == "deszcz" or pogoda == "śnieg":
    print("Weź parasol!")

# NOT - negacja warunku
zalogowany = False
if not zalogowany:
    print("Proszę się zalogować")

# Łączenie operatorów
print("\n=== Łączenie operatorów ===")
if wiek >= 18 and ma_prawo_jazdy and jest_trzezwy:
    print("Spełnia WSZYSTKIE warunki do prowadzenia")
else:
    print("NIE może prowadzić")
```
