## 3.1 Instrukcje warunkowe (if / elif / else)

### Podstawowa sk≈Çadnia

```python
# === PODSTAWOWA INSTRUKCJA IF ===

wiek = 20

if wiek >= 18:
    print("Jeste≈õ pe≈Çnoletni/a")
```

```python
# === IF / ELSE ===

wiek = 15

if wiek >= 18:
    print("Jeste≈õ pe≈Çnoletni/a")
else:
    print("Jeste≈õ niepe≈Çnoletni/a")
```

```python
# === IF / ELIF / ELSE ===
# elif = "else if" - pozwala sprawdzaƒá wiele warunk√≥w po kolei

temperatura = 25

if temperatura < 0:
    print("Mr√≥z! Ubierz siƒô ciep≈Ço!")
elif temperatura < 10:
    print("Zimno, we≈∫ kurtkƒô")
elif temperatura < 20:
    print("Ch≈Çodno, we≈∫ bluzƒô")
elif temperatura < 30:
    print("Przyjemnie ciep≈Ço")
else:
    print("Upa≈Ç! Pij du≈ºo wody!")

print(f"Aktualna temperatura: {temperatura}¬∞C")
```

### Operatory por√≥wnania

```python
# === OPERATORY POR√ìWNANIA ===

a = 10
b = 20

print("=== Operatory por√≥wnania ===")
print(f"a = {a}, b = {b}")
print(f"a == b (r√≥wne):          {a == b}")    # False
print(f"a != b (r√≥≈ºne):          {a != b}")    # True
print(f"a > b  (wiƒôksze):        {a > b}")     # False
print(f"a < b  (mniejsze):       {a < b}")     # True
print(f"a >= b (wiƒôksze-r√≥wne):  {a >= b}")    # False
print(f"a <= b (mniejsze-r√≥wne): {a <= b}")    # True
```

### Operatory logiczne (and, or, not)

```python
# === OPERATORY LOGICZNE ===

wiek = 25
ma_prawo_jazdy = True
jest_trzezwy = True

# AND - oba warunki muszƒÖ byƒá prawdziwe
if wiek >= 18 and ma_prawo_jazdy:
    print("Mo≈ºe prowadziƒá samoch√≥d")

# OR - wystarczy jeden warunek prawdziwy
pogoda = "deszcz"
if pogoda == "deszcz" or pogoda == "≈õnieg":
    print("We≈∫ parasol!")

# NOT - negacja warunku
zalogowany = False
if not zalogowany:
    print("Proszƒô siƒô zalogowaƒá")

# ≈ÅƒÖczenie operator√≥w
print("\n=== ≈ÅƒÖczenie operator√≥w ===")
if wiek >= 18 and ma_prawo_jazdy and jest_trzezwy:
    print("Spe≈Çnia WSZYSTKIE warunki do prowadzenia")
else:
    print("NIE mo≈ºe prowadziƒá")
```

### Zagnie≈ºd≈ºone warunki

```python
# === ZAGNIE≈ªD≈ªONE WARUNKI ===

wiek = 25
bilet = "ulgowy"
jest_studentem = True

print("=== System biletowy ===")

if wiek < 4:
    print("Przejazd darmowy (dziecko do 4 lat)")
else:
    if bilet == "ulgowy":
        if jest_studentem:
            cena = 1.50
            print(f"Bilet ulgowy studencki: {cena} z≈Ç")
        else:
            cena = 2.00
            print(f"Bilet ulgowy: {cena} z≈Ç")
    elif bilet == "normalny":
        cena = 3.50
        print(f"Bilet normalny: {cena} z≈Ç")
    else:
        print("Nieznany typ biletu!")
```

```python
# === WYRA≈ªENIE WARUNKOWE (TERNARY) ===
# sk≈Çadnia: warto≈õƒá_je≈õli_tak if warunek else warto≈õƒá_je≈õli_nie

wiek = 20

# Zamiast:
# if wiek >= 18:
#     status = "doros≈Çy"
# else:
#     status = "niepe≈Çnoletni"

# Mo≈ºna napisaƒá w jednej linii:
status = "doros≈Çy" if wiek >= 18 else "niepe≈Çnoletni"
print(f"Wiek: {wiek}, Status: {status}")

# Wiƒôcej przyk≈Çad√≥w
liczba = -5
typ = "dodatnia" if liczba > 0 else "ujemna" if liczba < 0 else "zero"
print(f"Liczba {liczba} jest {typ}")

# W praktyce - np. formatowanie tekstu
punkty = 1
tekst = f"Masz {punkty} punkt{'y' if punkty > 1 else ''}"
print(tekst)
```

### Sprawdzanie przynale≈ºno≈õci (in) i to≈ºsamo≈õci (is)

```python
# === OPERATOR 'in' ===

# Sprawdzanie czy element jest w kolekcji
owoce = ["jab≈Çko", "banan", "gruszka", "pomara≈Ñcza"]

szukany = "banan"
if szukany in owoce:
    print(f"'{szukany}' jest na li≈õcie owoc√≥w")

# Sprawdzanie w stringu
zdanie = "Python jest ≈õwietnym jƒôzykiem programowania"
if "Python" in zdanie:
    print("Zdanie zawiera s≈Çowo 'Python'")

# Sprawdzanie w s≈Çowniku (sprawdza klucze!)
oceny = {"matematyka": 5, "fizyka": 4, "chemia": 3}
if "matematyka" in oceny:
    print(f"Ocena z matematyki: {oceny['matematyka']}")

# === OPERATOR 'is' ===
# Sprawdza to≈ºsamo≈õƒá (czy to ten sam obiekt w pamiƒôci)
# NIE jest tym samym co == (por√≥wnanie warto≈õci)

a = None
if a is None:
    print("Zmienna 'a' jest None")

# is vs ==
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(f"\nx == y: {x == y}")  # True (te same warto≈õci)
print(f"x is y: {x is y}")    # False (r√≥≈ºne obiekty!)
print(f"x is z: {x is z}")    # True (ten sam obiekt)
```

### Prawdziwo≈õƒá warto≈õci (Truthy / Falsy)

```python
# === TRUTHY I FALSY VALUES ===
# Python traktuje niekt√≥re warto≈õci jako "fa≈Çszywe" (Falsy)

print("=== Warto≈õci FALSY (traktowane jako False) ===")

wartosci_falsy = [False, 0, 0.0, 0j, "", [], (), {}, set(), None, range(0)]

for val in wartosci_falsy:
    if not val:  # je≈õli warto≈õƒá jest falsy
        print(f"  {str(type(val).__name__):>10}: {str(val):<10} ‚Üí Falsy")

print("\n=== Warto≈õci TRUTHY (traktowane jako True) ===")

wartosci_truthy = [True, 1, -1, 3.14, "tekst", [1, 2], (1,), {"a": 1}, {1, 2}]

for val in wartosci_truthy:
    if val:  # je≈õli warto≈õƒá jest truthy
        print(f"  {str(type(val).__name__):>10}: {str(val):<10} ‚Üí Truthy")

# Praktyczne zastosowanie
print("\n=== Praktyczne zastosowanie ===")

lista = [1, 2, 3]

# Zamiast: if len(lista) > 0:
if lista:
    print(f"Lista nie jest pusta, ma {len(lista)} element√≥w")

nazwa = ""

# Zamiast: if nazwa != "":
if not nazwa:
    print("Nazwa jest pusta - podaj nazwƒô!")
```

### Praktyczny przyk≈Çad: Kalkulator BMI

```python
# === PRAKTYCZNY PRZYK≈ÅAD: KALKULATOR BMI ===

def oblicz_bmi(waga_kg, wzrost_cm):
    """Oblicza BMI i zwraca kategoriƒô."""
    
    # Walidacja danych
    if waga_kg <= 0 or wzrost_cm <= 0:
        return None, "B≈Çƒôdne dane! Waga i wzrost muszƒÖ byƒá dodatnie."
    
    if wzrost_cm < 50 or wzrost_cm > 250:
        return None, "Podaj realistyczny wzrost (50-250 cm)."
    
    # Obliczenie BMI
    wzrost_m = wzrost_cm / 100
    bmi = waga_kg / (wzrost_m ** 2)
    
    # Klasyfikacja
    if bmi < 16:
        kategoria = "wyg≈Çodzenie"
    elif bmi < 17:
        kategoria = "wychudzenie"
    elif bmi < 18.5:
        kategoria = "niedowaga"
    elif bmi < 25:
        kategoria = "waga prawid≈Çowa ‚úì"
    elif bmi < 30:
        kategoria = "nadwaga"
    elif bmi < 35:
        kategoria = "oty≈Ço≈õƒá I stopnia"
    elif bmi < 40:
        kategoria = "oty≈Ço≈õƒá II stopnia"
    else:
        kategoria = "oty≈Ço≈õƒá III stopnia"
    
    return bmi, kategoria

# Testowanie
print("=" * 45)
print("        KALKULATOR BMI")
print("=" * 45)

osoby = [
    ("Anna", 55, 165),
    ("Bartek", 95, 180),
    ("Celina", 45, 170),
    ("Daniel", 120, 175),
]

for imie, waga, wzrost in osoby:
    bmi, kategoria = oblicz_bmi(waga, wzrost)
    if bmi is not None:
        print(f"\n{imie}: waga={waga}kg, wzrost={wzrost}cm")
        print(f"  BMI = {bmi:.1f} ‚Üí {kategoria}")
```
